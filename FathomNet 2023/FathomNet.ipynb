```python
import pandas as pd
import json

# Specify the filename of the JSON file
filename = "/content/object_detection/eval.json"

# Load the JSON data
with open(filename, "r") as file:
    json_data = json.load(file)

# Extract image data from the JSON
image_data = []
for image in json_data["images"]:
    image_info = {
        "id": image["id"],
        "width": image["width"],
        "height": image["height"],
        "file_name": image["file_name"],
        "flickr_url": image["flickr_url"],
        "coco_url": image["coco_url"],
        "date_captured": image["date_captured"]
    }
    image_data.append(image_info)

# Create a DataFrame from the image data
df_eval = pd.DataFrame(image_data)

# Specify the filename for the CSV file
image_eval = "image_eval.csv"

# Convert the DataFrame to CSV format
df_eval.to_csv(image_eval, index=False)
```

```python
# Specify the filename of the JSON file
filename = "/content/object_detection/train.json"

# Load the JSON data
with open(filename, "r") as file:
    json_data = json.load(file)

# Extract image data from the JSON
image_data = []
for image in json_data["images"]:
    image_info = {
        "id": image["id"],
        "width": image["width"],
        "height": image["height"],
        "file_name": image["file_name"],
        "flickr_url": image["flickr_url"],
        "coco_url": image["coco_url"],
        "date_captured": image["date_captured"]
    }
    image_data.append(image_info)

# Create a DataFrame from the image data
df_object_detection = pd.DataFrame(image_data)

# Specify the filename for the CSV file
image_train = "image_train.csv"

# Convert the DataFrame to CSV format
df_object_detection.to_csv(image_train, index=False)
```
